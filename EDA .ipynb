```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("chotot_oto_database.csv")
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.info()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing
```

```python
# Giá
df["Giá_numeric"] = (
    df["Giá"]
    .str.replace(".", "", regex=False)
    .str.replace(" đ", "", regex=False)
    .astype(float)
)

# Numeric columns
df["Năm sản xuất"] = pd.to_numeric(df["Năm sản xuất"], errors="coerce")
df["Số Km đã đi"] = pd.to_numeric(df["Số Km đã đi"], errors="coerce")

# Feature engineering
CURRENT_YEAR = 2025
df["Tuổi xe"] = CURRENT_YEAR - df["Năm sản xuất"]

df[["Giá_numeric", "Năm sản xuất", "Số Km đã đi", "Tuổi xe"]].describe()
```

```python
# HISTOGRAM GIÁ
plt.figure(figsize=(8,5))
plt.hist(df["Giá_numeric"].dropna(), bins=60, color="cornflowerblue", edgecolor="black")
plt.title("Phân phối giá ô tô")
plt.xlabel("Giá (VND)")
plt.ylabel("Số lượng")
plt.show()
```

```python
# HISTOGRAM LOG(GIÁ)
plt.figure(figsize=(8,5))
plt.hist(np.log10(df["Giá_numeric"].dropna()), bins=50, color="orange")
plt.title("Phân phối log10(Giá)")
plt.xlabel("log10(Giá)")
plt.ylabel("Số lượng")
plt.show()
```

```python
# PHÂN PHỐI NĂM SẢN XUẤT
plt.figure(figsize=(8,5))
plt.hist(df["Năm sản xuất"].dropna(), bins=25, color="seagreen")
plt.title("Phân phối năm sản xuất")
plt.xlabel("Năm")
plt.ylabel("Số lượng")
plt.show()
```

```python
# PHÂN PHỐI TUỔI XE
plt.figure(figsize=(8,5))
plt.hist(df["Tuổi xe"].dropna(), bins=25, color="mediumpurple")
plt.title("Phân phối tuổi xe")
plt.xlabel("Tuổi xe")
plt.ylabel("Số lượng")
plt.show()
```

```python
# PHÂN PHỐI SỐ KM
plt.figure(figsize=(8,5))
plt.hist(df["Số Km đã đi"].dropna(), bins=60, color="salmon")
plt.title("Phân phối số km đã đi")
plt.xlabel("Km")
plt.ylabel("Số lượng")
plt.show()
```

```python
# TOP 10 HÃNG XE
df["Hãng xe"].value_counts().head(10).plot(kind="bar", figsize=(8,5))
plt.title("Top 10 hãng xe phổ biến")
plt.xlabel("Hãng xe")
plt.ylabel("Số lượng")
plt.show()
```

```python
# KIỂU DÁNG XE
df["Kiểu dáng"].value_counts().head(8).plot(kind="bar", figsize=(8,5))
plt.title("Phân bố kiểu dáng xe")
plt.xlabel("Kiểu dáng")
plt.ylabel("Số lượng")
plt.show()
```

```python
# BIỂU ĐỒ TRÒN KIỂU DÁNG
body = df["Kiểu dáng"].value_counts().head(6)

plt.figure(figsize=(6,6))
plt.pie(body, labels=body.index, autopct="%1.1f%%", startangle=90)
plt.title("Tỷ trọng kiểu dáng xe")
plt.show()
```

```python
# NHIÊN LIỆU
df["Nhiên liệu"].value_counts().plot(kind="bar", figsize=(7,5))
plt.title("Phân bố nhiên liệu")
plt.xlabel("Nhiên liệu")
plt.ylabel("Số lượng")
plt.show()
```

```python
# HỘP SỐ
df["Hộp số"].value_counts().plot(kind="bar", figsize=(7,5))
plt.title("Phân bố hộp số")
plt.xlabel("Hộp số")
plt.ylabel("Số lượng")
plt.show()
```

```python
# GIÁ vs NĂM SẢN XUẤT
plt.figure(figsize=(8,5))
plt.scatter(df["Năm sản xuất"], df["Giá_numeric"], alpha=0.6)
plt.title("Giá vs Năm sản xuất")
plt.xlabel("Năm sản xuất")
plt.ylabel("Giá")
plt.show()
```

```python
# GIÁ vs KM
plt.figure(figsize=(8,5))
plt.scatter(df["Số Km đã đi"], df["Giá_numeric"], alpha=0.6)
plt.title("Giá vs Số km đã đi")
plt.xlabel("Km")
plt.ylabel("Giá")
plt.show()
```

```python
# GIÁ vs TUỔI XE (COLOR)
plt.figure(figsize=(8,5))
plt.scatter(
    df["Tuổi xe"],
    df["Giá_numeric"],
    c=df["Tuổi xe"],
    alpha=0.7
)
plt.colorbar(label="Tuổi xe")
plt.title("Giá theo tuổi xe")
plt.xlabel("Tuổi xe")
plt.ylabel("Giá")
plt.show()
```

```python
# GIÁ TRUNG BÌNH THEO NĂM (LINE)
df.groupby("Năm sản xuất")["Giá_numeric"].mean().plot(
    figsize=(8,5), marker="o"
)
plt.title("Giá trung bình theo năm sản xuất")
plt.xlabel("Năm")
plt.ylabel("Giá trung bình")
plt.show()
```

```python
# BOXPLOT GIÁ THEO HÃNG (TOP 5
top5 = df["Hãng xe"].value_counts().head(5).index
data = [df[df["Hãng xe"] == b]["Giá_numeric"] for b in top5]

plt.figure(figsize=(8,5))
plt.boxplot(data, labels=top5)
plt.title("So sánh giá theo hãng")
plt.ylabel("Giá")
plt.show()
```

```python
# MA TRẬN TƯƠNG QUAN
corr = df[["Giá_numeric", "Năm sản xuất", "Số Km đã đi", "Tuổi xe"]].corr()

plt.figure(figsize=(6,5))
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Ma trận tương quan")
plt.show()
```

```python
# TOP 10 XE ĐẮT NHẤT
top_price = df.sort_values("Giá_numeric", ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.barh(top_price["Tên sản phẩm"], top_price["Giá_numeric"])
plt.title("Top 10 xe có giá cao nhất")
plt.xlabel("Giá")
plt.gca().invert_yaxis()
plt.show()
```
