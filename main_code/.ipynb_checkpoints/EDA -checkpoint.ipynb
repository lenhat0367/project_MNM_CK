```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("chotot_oto_database.csv")
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
# Giá
df["Giá_numeric"] = (
    df["Giá"]
    .str.replace(".", "", regex=False)
    .str.replace(" đ", "", regex=False)
    .astype(float)
)

# Numeric columns
df["Năm sản xuất"] = pd.to_numeric(df["Năm sản xuất"], errors="coerce")
df["Số Km đã đi"] = pd.to_numeric(df["Số Km đã đi"], errors="coerce")

# Feature engineering
CURRENT_YEAR = 2025
df["Tuổi xe"] = CURRENT_YEAR - df["Năm sản xuất"]

df[["Giá_numeric", "Năm sản xuất", "Số Km đã đi", "Tuổi xe"]].describe()
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.info()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing
```

```python
# HISTOGRAM GIÁ
from matplotlib.ticker import FuncFormatter
def format_vnd(x, pos):
    if x >= 1_000_000_000:
        return f"{x/1_000_000_000:.1f} tỷ"
    elif x >= 1_000_000:
        return f"{int(x/1_000_000)} triệu"
    else:
        return f"{int(x)}"

plt.figure(figsize=(8,5))
plt.hist(
    df["Giá_numeric"].dropna(),
    bins=60,
    color="cornflowerblue",
    edgecolor="black"
)

plt.title("Phân phối giá ô tô")
plt.xlabel("Giá (VND)")
plt.ylabel("Số lượng")

plt.xlim(0, 3_000_000_000)
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_vnd))

plt.show()

```

```python
# HISTOGRAM LOG(GIÁ)
log_price = np.log10(df["Giá_numeric"].dropna())

plt.hist(log_price, bins=50, color="orange", edgecolor="black")

plt.title("Phân phối Giá ô tô")
plt.xlabel("Giá (VND, )")
plt.ylabel("Số lượng")

# Formatter: log10 -> giá gốc VND
def log_vnd_formatter(x, pos):
    price = 10 ** x
    if price >= 1_000_000_000:
        return f"{price/1_000_000_000:.1f} tỷ"
    else:
        return f"{int(price/1_000_000)} triệu"

plt.gca().xaxis.set_major_formatter(FuncFormatter(log_vnd_formatter))

plt.show()

```

```python
# PHÂN PHỐI NĂM SẢN XUẤT
plt.figure(figsize=(8,5))
plt.hist(df["Năm sản xuất"].dropna(), bins=25, color="seagreen")
plt.title("Phân phối năm sản xuất")
plt.xlabel("Năm")
plt.ylabel("Số lượng")
plt.show()
```

```python
# PHÂN PHỐI TUỔI XE
plt.figure(figsize=(8,5))
plt.hist(df["Tuổi xe"].dropna(), bins=25, color="mediumpurple")
plt.title("Phân phối tuổi xe")
plt.xlabel("Tuổi xe")
plt.ylabel("Số lượng")
plt.show()
```

```python
# PHÂN PHỐI SỐ KM
plt.hist(
    df["Số Km đã đi"].dropna(),
    bins=60,
    color="salmon",
    edgecolor="black"
)

plt.title("Phân phối số km đã đi")
plt.xlabel("Số km đã đi")
plt.ylabel("Số lượng")

# Formatter km -> nghìn km
def km_formatter(x, pos):
    if x >= 1_000:
        return f"{int(x/1_000)}k km"
    else:
        return f"{int(x)}"

plt.gca().xaxis.set_major_formatter(FuncFormatter(km_formatter))

plt.show()

```

```python
# TOP 10 HÃNG XE
top10 = df["Hãng xe"].value_counts().head(10)

plt.figure(figsize=(8,5))
top10.sort_values().plot(kind="barh", color="steelblue", edgecolor="black")

plt.title("Top 10 hãng xe phổ biến")
plt.xlabel("Số lượng")
plt.ylabel("Hãng xe")
plt.show()
```

```python
# BIỂU ĐỒ TRÒN KIỂU DÁNG
body = df["Kiểu dáng"].value_counts().head(6)

plt.figure(figsize=(6,6))
plt.pie(body, labels=body.index, autopct="%1.1f%%", startangle=90)
plt.title("Tỷ trọng kiểu dáng xe")
plt.show()
```

```python
# NHIÊN LIỆU
df["Nhiên liệu"].value_counts().plot(kind="bar", figsize=(7,5))
plt.title("Phân bố nhiên liệu")
plt.xlabel("Nhiên liệu")
plt.ylabel("Số lượng")
plt.xticks(rotation=0)
plt.show()
```

```python
# Chọn các cột cần tính tương quan
corr = df[[
    "Giá_numeric",
    "Năm sản xuất",
    "Số Km đã đi",
    "Tuổi xe"
]].corr()

plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,              # HIỂN THỊ SỐ
    fmt=".2f",               # 2 chữ số thập phân
    cmap="coolwarm",         # giống hình mẫu
    linewidths=0.5,
    square=True,
    cbar=True
)

plt.title("Ma trận tương quan giữa các chỉ số")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
